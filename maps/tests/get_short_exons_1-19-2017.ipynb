```python
df = pd.read_table('/home/elvannostrand/data/ENCODE/RNAseq/Brent_RNASEQlist_fromDCC.tsv.file_accessions_HepG2.hg19_V19.20161207.tsv.strict_CE_0.8',
                  names=['event','lo','exon','hi'])
def get_len(row):
    start, end = row['exon'].strip().split('-')
    return int(end) - int(start)

df['len'] = df.apply(get_len,axis=1)
df.sort_values(by='len')
```
